````python
import soundata

dataset = soundata.initialize('urbansound8k')
dataset.download()  # download the dataset
dataset.validate()  # validate that all the expected files are there

example_clip = dataset.choice_clip()  # choose a random example clip
print(example_clip)  # see the available data
````

Let's train a deep learning model for classification. We will use a simple Multilayer Perceptron (MLP) with TensorFlow/Keras.

First, let's load all the clip metadata and then extract audio features (MFCCs) for each clip. MFCCs are widely used in audio signal processing to represent the spectral characteristics of sounds. This step will create the `X` (features) and `y` (raw labels) variables.

````python
import librosa
import numpy as np
import pandas as pd
import os

# Get all clips from the dataset
# Corrected: Use dataset.clip_ids to get a list of all clip IDs
all_clip_ids = dataset.clip_ids

# Initialize lists to store features and labels
features = []
labels = []

print(f"Extracting features from {len(all_clip_ids)} audio clips...")

for clip_id in all_clip_ids:
    clip = dataset.clip(clip_id)

    # Load audio data
    try:
        y_audio, sr_audio = librosa.load(clip.audio_path, sr=None) # sr=None to preserve original sample rate
    except Exception as e:
        print(f"Error loading audio for clip {clip_id}: {e}")
        continue

    # Extract MFCCs (Mel-frequency cepstral coefficients)
    # MFCCs are commonly used features for audio classification
    mfccs = librosa.feature.mfcc(y=y_audio, sr=sr_audio, n_mfcc=40) # Using 40 MFCCs

    # Aggregate MFCCs (e.g., mean and variance over time)
    mfccs_processed = np.mean(mfccs.T, axis=0)

    features.append(mfccs_processed)
    labels.append(clip.class_label) # The target label for classification

# Convert to numpy arrays
X = np.array(features)
y = np.array(labels)

print("Feature extraction complete.")
print(f"Shape of features (X): {X.shape}")
print(f"Shape of labels (y): {y.shape}")
````

Next, we'll encode the string labels into numerical format using `LabelEncoder`. This will create `y_encoded` and `le`, which are necessary for model training.

````python
from sklearn.preprocessing import LabelEncoder

# Encode labels to numerical format
le = LabelEncoder()
y_encoded = le.fit_transform(y)

print(f"Original labels: {np.unique(y)}")
print(f"Encoded labels: {np.unique(y_encoded)}")
print(f"Label mapping: {list(zip(le.classes_, le.transform(le.classes_)))}")
````

````python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split # Added import
from sklearn.metrics import classification_report # Added import

# Convert labels to one-hot encoding for Keras
y_one_hot = to_categorical(y_encoded)

# Get number of classes
num_classes = y_one_hot.shape[1]

# Split the data into training and testing sets for deep learning
X_train_dl, X_test_dl, y_train_dl, y_test_dl = train_test_split(X, y_one_hot, test_size=0.2, random_state=42)

print(f"Deep Learning Training set size: {X_train_dl.shape[0]} samples")
print(f"Deep Learning Test set size: {X_test_dl.shape[0]} samples")

# Define the Keras model
model_dl = Sequential([
    Dense(256, activation='relu', input_shape=(X_train_dl.shape[1],)),
    Dropout(0.5),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(num_classes, activation='softmax')
])

# Compile the model
model_dl.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Print model summary
model_dl.summary()

print("Training the Deep Learning model...")
# Train the model
history = model_dl.fit(
    X_train_dl, y_train_dl,
    validation_data=(X_test_dl, y_test_dl),
    epochs=50, # You might want to adjust the number of epochs
    batch_size=32,
    verbose=1
)
print("Deep Learning model training complete.")

# Evaluate the model on the test set
loss, accuracy_dl = model_dl.evaluate(X_test_dl, y_test_dl, verbose=0)
print(f"\nDeep Learning Model Accuracy: {accuracy_dl:.4f}")

# Optional: Display classification report for deep learning model
# Convert one-hot predictions back to class labels
y_pred_dl = np.argmax(model_dl.predict(X_test_dl), axis=1)
y_test_labels = np.argmax(y_test_dl, axis=1)

print("\nDeep Learning Classification Report:")
print(classification_report(y_test_labels, y_pred_dl, target_names=le.classes_))
````

### Hyperparameter Tuning with Keras Tuner

To systematically explore different hyperparameters (like the number of layers, units per layer, dropout rates, and learning rate), we'll use `Keras Tuner`. This library helps automate the process of finding the optimal set of hyperparameters for your model.

````python
# Install Keras Tuner
!pip install -q keras-tuner
````

````python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import keras_tuner as kt
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
import numpy as np

# Ensure y_encoded and num_classes are available (from previous steps)
if 'y_encoded' not in globals():
    print("y_encoded not found. Please ensure previous cells for label encoding have been run.")
    raise SystemExit

y_one_hot = to_categorical(y_encoded)
num_classes = y_one_hot.shape[1]

# Ensure X is available
if 'X' not in globals():
    print("X (features) not found. Please ensure feature extraction cells have been run.")
    raise SystemExit

# Split data (if not already split)
if 'X_train_dl' not in globals() or 'X_test_dl' not in globals():
    X_train_dl, X_test_dl, y_train_dl, y_test_dl = train_test_split(X, y_one_hot, test_size=0.2, random_state=42)
    print("Data split into training and testing sets for Keras Tuner.")

# Define the model-building function for Keras Tuner
def build_tunable_model(hp):
    model = keras.Sequential()
    model.add(layers.InputLayer(input_shape=(X_train_dl.shape[1],)))

    # Tune the number of hidden layers (1 to 3)
    for i in range(hp.Int('num_layers', 1, 3)):
        model.add(layers.Dense(
            units=hp.Int(f'units_{i}', min_value=32, max_value=512, step=32),
            activation='relu'
        ))
        # Tune dropout rate for each hidden layer
        model.add(layers.Dropout(hp.Float(f'dropout_{i}', min_value=0.2, max_value=0.5, step=0.1)))

    model.add(layers.Dense(num_classes, activation='softmax'))

    # Tune the learning rate for the optimizer
    hp_learning_rate = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])
    optimizer = keras.optimizers.Adam(learning_rate=hp_learning_rate)

    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# Instantiate the tuner
tuner = kt.RandomSearch(
    build_tunable_model,
    objective='val_accuracy',
    max_trials=10,  # Number of hyperparameter combinations to try
    executions_per_trial=2, # Number of models to train for each trial to reduce variance
    directory='keras_tuner_dir', # Directory to store results
    project_name='urbansound8k_tuning', # Project name
    overwrite=True # Overwrite previous results if they exist
)

print("Starting hyperparameter search...")
# Run the hyperparameter search
tuner.search(
    X_train_dl, y_train_dl,
    epochs=10,  # Number of epochs for each trial (can be smaller for quick search)
    validation_data=(X_test_dl, y_test_dl),
    verbose=1
)
print("Hyperparameter search complete.")

# Get the best hyperparameters
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
print("\nBest hyperparameters found:")
print(best_hps.values)

# Build the best model
best_model = tuner.get_best_models(num_models=1)[0]

print("\nTraining the final model with best hyperparameters...")
# Retrain the best model with more epochs (optional, but good practice)
history_tuned = best_model.fit(
    X_train_dl, y_train_dl,
    epochs=50,  # Train for more epochs with the best model
    validation_data=(X_test_dl, y_test_dl),
    verbose=1
)

# Evaluate the best model on the test set
loss_tuned, accuracy_tuned = best_model.evaluate(X_test_dl, y_test_dl, verbose=0)
print(f"\nTunable Model Accuracy: {accuracy_tuned:.4f}")

# Display classification report for the tuned model
y_pred_tuned = np.argmax(best_model.predict(X_test_dl), axis=1)
y_test_labels = np.argmax(y_test_dl, axis=1)

print("\nTunable Model Classification Report:")
print(classification_report(y_test_labels, y_pred_tuned, target_names=le.classes_))
````

### Preparing the Keras Model for a Frontend Application

To use your trained Keras model in a web frontend application, you typically need to convert it into a format that can be consumed by JavaScript. The two most common approaches are using **TensorFlow.js** or converting to **ONNX**.

#### 1. Using TensorFlow.js for JavaScript-based Frontends

TensorFlow.js is a library that allows you to develop machine learning models in JavaScript, and also to convert and run existing TensorFlow/Keras models directly in the browser or on Node.js. This is often the most straightforward path for web applications.

**Steps:**
1.  **Convert the Keras model:** You'll use the `tensorflowjs_converter` tool to convert your saved Keras model into the TensorFlow.js format.

    ```bash
    !pip install tensorflowjs
    !tensorflowjs_converter --input_format keras /content/drive/My\ Drive/urbansound8k_model/urbansound8k_mlp_model.keras /content/drive/My\ Drive/urbansound8k_model/tfjs_model
    ```

    This will create a directory (`tfjs_model` in this example) containing a `model.json` file and a set of shard files (e.g., `group1-shard1of1.bin`).

2.  **Load and use in JavaScript:** In your frontend JavaScript code, you can then load and use this model:

    ```javascript
    import * as tf from '@tensorflow/tfjs';

    async function loadModelAndPredict() {
        const model = await tf.loadLayersModel('https://your-domain.com/path/to/tfjs_model/model.json');
        // or if running locally:
        // const model = await tf.loadLayersModel('./tfjs_model/model.json');

        // Prepare input data (e.g., MFCCs from an audio file processed in JavaScript)
        const inputTensor = tf.tensor2d([[/* your 40 MFCC values here */]]);

        const predictions = model.predict(inputTensor);
        const predictedClass = predictions.argMax(-1).dataSync()[0];
        // You would also need to load your label_encoder.json or map the index back to a label
        console.log('Predicted class index:', predictedClass);
    }

    loadModelAndPredict();
    ```

3.  **Handle the LabelEncoder:** You would also need to save your `label_encoder.pkl` as a JSON file and load it in JavaScript to map the numerical predictions back to human-readable class labels.

#### 2. Converting to ONNX

ONNX (Open Neural Network Exchange) is an open format designed to represent machine learning models. It allows interoperability between different ML frameworks. You can convert your Keras model to ONNX, and then use a JavaScript-compatible ONNX runtime (like ONNX Runtime Web) in your frontend.

**Steps:**
1.  **Install `onnx` and `tf2onnx`:**

    ```bash
    !pip install onnx tf2onnx
    ```

2.  **Convert Keras to ONNX:**

    ```python
    import tensorflow as tf
    import tf2onnx
    import onnx
    import os

    # Load your Keras model
    model_path = '/content/drive/My Drive/urbansound8k_model/urbansound8k_mlp_model.keras'
    keras_model = tf.keras.models.load_model(model_path)

    # Define input signature (important for conversion)
    input_signature = [
        tf.TensorSpec(keras_model.inputs[0].shape, keras_model.inputs[0].dtype, name="input_features")
    ]

    onnx_model_path = os.path.join('/content/drive/My Drive/urbansound8k_model', 'urbansound8k_mlp_model.onnx')
    model_proto, _ = tf2onnx.convert.from_keras(keras_model, input_signature, opset=13, output_path=onnx_model_path)
    print(f"ONNX model saved to: {onnx_model_path}")
    ```

3.  **Load and use with ONNX Runtime Web:** In your frontend, you would use `@microsoft/onnxruntime-web` to load and run the ONNX model.

    ```javascript
    import * as ort from 'onnxruntime-web';

    async function loadOnnxModelAndPredict() {
        const session = await ort.InferenceSession.create('https://your-domain.com/path/to/urbansound8k_mlp_model.onnx');
    ```
